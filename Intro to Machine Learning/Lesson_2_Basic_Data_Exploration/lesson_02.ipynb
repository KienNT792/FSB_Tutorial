## Interpreting Data Description – Hiểu kết quả mô tả dữ liệu

Khi dùng Pandas, phương thức `describe()` trả về **8 số liệu** cho mỗi cột trong dataset gốc của bạn:

1. **count** – số lượng giá trị không bị thiếu (non-missing).
   - Missing values có thể xuất hiện vì nhiều lý do. Ví dụ: kích thước phòng ngủ thứ 2 sẽ không được ghi khi khảo sát nhà chỉ có 1 phòng ngủ.

2. **mean** – giá trị trung bình (average).

3. **std** – độ lệch chuẩn (standard deviation), đo mức độ phân tán của các giá trị.

4. **min, 25%, 50%, 75%, max** – các giá trị theo thứ tự từ thấp đến cao:
   - **min** – giá trị nhỏ nhất.
   - **25%** – giá trị tại 25th percentile, lớn hơn 25% các giá trị và nhỏ hơn 75% còn lại.
   - **50%** – giá trị trung vị (median), lớn hơn 50% các giá trị còn lại.
   - **75%** – giá trị tại 75th percentile, lớn hơn 75% các giá trị còn lại.
   - **max** – giá trị lớn nhất.

Bạn có thể hình dung việc sắp xếp cột từ nhỏ đến lớn để hiểu các giá trị này.
